```python
# Import pairwise2 module
from Bio import pairwise2

# Import format_alignment method
from Bio.pairwise2 import format_alignment

# Define two sequences to be aligned
X = "She said, It hath reached me, O auspicious King, that the Wazir', 'stood up to him and welcoming him said,"
Y = "I have heard, O auspicious king, that the vizier got up to greet him, saying:"

# Get a list of the global alignments between the two sequences 
alignments = pairwise2.align.globalxx(X, Y)
#print(alignments)

# Use format_alignment method to format the alignments in the list
for a in alignments:
    print(format_alignment(*a))
```
